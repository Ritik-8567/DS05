```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('/content/Health_Familiar_Statistics_2017_Table_E.29.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
correlation = df[['Women  Tobacco (%) - NFHS-4 ', 'Men Tobacco (%) - NFHS-4 - Total']].corr()
correlation
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
avg_women_tobacco = df['Women  Tobacco (%) - NFHS-4 '].mean()
avg_men_tobacco = df['Men Tobacco (%) - NFHS-4 - Total'].mean()
```

```python
avg_women_tobacco
```

```python
avg_men_tobacco
```

```python
plt.bar(['Women', 'Men'], [avg_women_tobacco, avg_men_tobacco], alpha=0.7)
plt.xlabel('Gender')
plt.ylabel('Average Tobacco Usage (%)')
plt.title('Comparison of Tobacco Usage between Men and Women')
plt.show()
```

```python
# Plotting the distribution of "Women who use any kind of Tobacco (%)-NFHS-4 Total"
plt.hist(df['Women  Tobacco (%) - NFHS-4 '], bins=10)
plt.xlabel('Women who use any kind of Tobacco (%) - NFHS-4 - Total')
plt.ylabel('Frequency')
plt.title('Distribution of Women who use any kind of Tobacco (%) - NFHS-4 - Total')
plt.show()
```

```python
plt.hist(df['Men Tobacco (%) - NFHS-4 - Total'], bins=10)
plt.xlabel('Men who use any kind of Tobacco (%) - NFHS-4 - Total')
plt.ylabel('Frequency')
plt.title('Distribution of Men who use any kind of Tobacco (%) - NFHS-4 - Total')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Women  Tobacco (%) - NFHS-4 '], bins=20, alpha=0.7, label='Women - NFHS-4')
plt.hist(df['Men Tobacco (%) - NFHS-4 - Total'], bins=20, alpha=0.7, label='Men - NFHS-4')
plt.xlabel('Tobacco Usage (%)')
plt.ylabel('Frequency')
plt.title('Distribution of Tobacco Usage')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
grouped_data = df.groupby('Men Tobacco (%) - NFHS-4 - Urban')['Men Tobacco (%) - NFHS-4 - Total'].mean()

# Plot the results
grouped_data.plot(kind='bar')
plt.xlabel('Urban/Rural')
plt.ylabel('Average Men Tobacco Usage (%)')
plt.title('Average Men Tobacco Usage based on Urban/Rural Status')
plt.show()
```

```python
grouped_data = df.groupby('Women Tobacco (%) - NFHS-4 - Urban')['Women  Tobacco (%) - NFHS-4 '].mean()

# Plot the results
grouped_data.plot(kind='bar')
plt.xlabel('Urban/Rural')
plt.ylabel('Average Women Tobacco Usage (%)')
plt.title('Average Women Tobacco Usage based on Urban/Rural Status')
plt.show()
```
